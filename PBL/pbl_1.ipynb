```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler

# ==========================================
# 1. 데이터 로드 및 초기 확인
# ==========================================
# 당뇨병 데이터셋 불러오기
df = pd.read_csv('diabetes.csv')

# ==========================================
# 2. 결측치(Missing Value) 처리
# ==========================================
# 0으로 입력된 유효하지 않은 데이터들을 NaN(결측치)으로 변환
cols_to_nan = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']
df[cols_to_nan] = df[cols_to_nan].replace(0, np.nan)

# [EDA 요구사항] 결측치 처리 직후의 상태 확인을 위해 NaN 개수 출력
print("=== [EDA] 각 컬럼별 결측치(NaN) 개수 ===")
print(df.isnull().sum())
print("-" * 50)

# 각 컬럼의 '전체 평균값'을 계산하여 결측치(NaN)를 대체
df[cols_to_nan] = df[cols_to_nan].fillna(df[cols_to_nan].mean())

# ==========================================
# 3. 이상치(Outlier) 처리
# ==========================================
# 상위 1%를 벗어나는 극단적인 값을 처리할 컬럼 지정
first = ['SkinThickness', 'Insulin']

for col in first:
    # 해당 컬럼의 상위 99%에 해당하는 분위수(기준점) 계산
    q99 = df[col].quantile(0.99)
    
    # 결측치가 채워진 전체 데이터의 평균값 계산
    col_mean = df[col].mean()
    
    # 기준점(q99)을 초과하는 이상치 데이터들을 전체 평균값으로 대체
    df.loc[df[col] > q99, col] = col_mean

# ==========================================
# 4. 데이터 스케일링 (Feature Scaling)
# ==========================================
# Age 컬럼에 대해 0과 1 사이로 정규화하는 MinMaxScaler 적용
scaler = MinMaxScaler()
df['Age'] = scaler.fit_transform(df[['Age']])

# ==========================================
# 5. 최종 결과 확인 및 EDA 출력
# ==========================================
# [EDA 요구사항 1] 당뇨병 여부(Outcome)에 따른 포도당(Glucose)의 평균 수치 출력
print("=== [EDA] 당뇨병 발병 여부(Outcome)별 Glucose 평균 ===")
print(df.groupby('Outcome')['Glucose'].mean())
print("-" * 50)

# [EDA 요구사항 2] 모든 전처리가 완료된 후 데이터프레임의 상위 5개 행 출력
print("=== [EDA] 전처리 완료 후 데이터프레임 상위 5개 행 ===")
print(df.head(5))
print("-" * 50)
```
