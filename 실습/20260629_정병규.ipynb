#### 대기업 모바일 게임 제작 통계

```python
import pandas as pd
import matplotlib.pyplot as plt

# 한글 폰트 설정
plt.rc('font',family='Malgun Gothic') # 맑은 고딕

# "-" 부호 깨짐 방지 
plt.rc('axes', unicode_minus=False)
```

```python
df = pd.read_csv('문화_게임물제작업.csv', encoding="cp949")
```

```python
df.info()
```

```python
df.columns
```

```python
df_1 = df[['영업상태명','사업장명','제작취급품목내용']]
df_1
```

```python
df_2 = df_1[df_1['영업상태명'].str.contains('영업')]
df_2
```

```python
df_3 = df_2[df_2['사업장명'].str.contains('넥슨|크래프톤|넷마블|카카오')]
df_3
```

```python
df_4 = df_3[df_3['제작취급품목내용'].str.contains('모바일')]
df_4
```

```python
df_5 = df_4.drop('영업상태명', axis=1)
```

```python
df_5['기업그룹'] = df_5['사업장명'].str.extract(f'(넷마블|넥슨|카카오)') 
df_6 = df_5.groupby('기업그룹').size().reset_index(name='제작 개수')
df_6
```

```python
plt.bar(df_6['기업그룹'],df_6['제작 개수'], color='lightgreen')
plt.xlabel('기업그룹')
plt.ylabel('모바일 게임 제작 개수')
plt.yticks(range(1,5))
plt.grid(linestyle='--',alpha = 0.5)
plt.show()
```
