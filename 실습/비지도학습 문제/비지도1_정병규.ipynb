비지도학습 문제 1

```python
from sklearn.ensemble import IsolationForest
import numpy as np

# 1. 미션용 가상 센서 데이터 생성 (제시된 코드)
rng = np.random.RandomState(100)
X_sensor_normal = 0.5 * rng.randn(300, 5)
X_sensor_faulty = rng.uniform(low=-5, high=5, size=(25, 5))
X_sensor = np.vstack([X_sensor_normal, X_sensor_faulty])

# 2. 하이퍼파라미터 조건에 맞춰 Isolation Forest 모델 선언 및 학습
# 조건: contamination=0.08, random_state=42
model = IsolationForest(
    contamination=0.08,
    random_state=42
)
model.fit(X_sensor)
# 3. 전체 센서 데이터에 대한 정상/이상 판정 결과 저장
# 정상은 1, 이상치는 -1로 반환됩니다.
outlier_preds = model.predict(X_sensor)

# 4. 이상치로 최종 분류된 샘플(레이블이 -1)의 개수 계산 및 출력
anomaly_count = np.sum(outlier_preds == -1)

print(f"이상치로 분류된 샘플 개수: {anomaly_count}")
```
