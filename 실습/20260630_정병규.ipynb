도서 목록 데이터 추출 및 구조화

```python
from bs4 import BeautifulSoup as bs

html_books = """
<div class='library'>
<div class='book' id='b-101'>
<h3 class='book-title'>딥러닝 텐서플로 마스터</h3>
<p class='book-author'>김작가</p>
</div>
<div class='book' id='b-102'>
<h3 class='book-title'>이것이 진짜 R 프로그래밍</h3>
<p class='book-author'>박통계</p>
</div>
<div class='book' id='b-103'>
<h3 class='book-title'>자연어 처리 입문</h3>
<p class='book-author'>최이과</p>
</div>
</div>
"""

soup = bs(html_books, 'html.parser')
books_info = []

books = soup.select('.book')

# ==========================================
# TODO: Mission 1 코드를 작성하세요.
# ==========================================


for book in books:
    title = book.select_one('.book-title').text
    author = book.select_one('.book-author').text
    books_info.append({
        'title': title,
        'author': author
    })


print("=== 추출된 도서 목록 ===")
print(books_info)
```

```python
from bs4 import BeautifulSoup as bs

html_news = """
<ul class='news-section'>
<li class='news-item'><a href='https://news.example.com/ai-1'>인공지능 규제법 통과, IT 업계 파장</a></li>
<li class='news-item'><a href='https://news.example.com/dev-2'>파이썬 3.12 릴리즈, 무엇이 달라졌나?</a></li>
<li class='news-item'><a href='https://news.example.com/sec-3'>오픈소스 라이브러리 보안 취약점 경고</a></li>
</ul>
"""
soup = bs(html_news, 'html.parser')

news = soup.select('.news-item')
news_list = []

for new in news:
    headline = new.select_one('a').text
    url = new.select_one('a')['href']
    news_list.append({'headline': headline, 'url': url})
# ==========================================
# TODO: Mission 2 코드를 작성하세요.
# ==========================================

print("=== 수집된 뉴스 헤드라인 ===")
print(news_list)
```

SQLite 데이터 조작 (CRUD)

```python
import sqlite3

conn = sqlite3.connect('user_data.db')

sql_create = '''
CREATE TABLE IF NOT EXISTS users (
id INTEGER PRIMARY KEY AUTOINCREMENT,
name TEXT NOT NULL,
email TEXT NOT NULL UNIQUE,
level TEXT
)
'''

cursor = conn.cursor()
cursor.execute(sql_create)
conn.commit()
```

수집된 데이터프레임의 DB 테이블 저장 및 복원

```python
import pandas as pd
import sqlite3

conn = sqlite3.connect('users.db')
sql_create = '''
CREATE TABLE IF NOT EXISTS members (
member_id TEXT PRIMARY KEY,
name TEXT,
score INTEGER
)
'''
cursor = conn.cursor()
cursor.execute(sql_create)


# 실습 데이터프레임 생성
df_members = pd.DataFrame({
'member_id': ['M01', 'M02', 'M03'],
'name': ['이민수', '김영희', '박준형'],
'score': [85, 95, 78]
})

df_members.to_sql('members', conn, if_exists='append', index=False)

conn.close()

# ==========================================
# TODO: Mission 2 코드를 작성하세요.
# ==========================================

conn_read = sqlite3.connect('users.db')

df_read = pd.read_sql_query("SELECT * FROM members", conn_read)

conn_read.close()

print("=== DB 복원 데이터프레임 ===")
print(df_read)
```
