```python
import statistics
data = [10, 15, 10, 20, 30, 10, 40]
# 평균값구하기
mean_value = statistics.mean(data)
print("평균:", mean_value)
# 중앙값구하기
median_value = statistics.median(data)
print("중앙값:", median_value)
# 최빈값구하기
mode_value = statistics.mode(data)
print("최빈값:", mode_value)
# 여러 최빈값구하기
from statistics import multimode
modes = multimode([1, 2, 2, 3, 3])
print("여러 최빈값:", modes) # 결과: [2, 3]
```

```python
import statistics
data = [10, 15, 10, 20, 30, 10, 40]
# 분산 구하기
var_value = statistics.variance(data)
print("분산:", var_value)
# 표준편차구하기
std_dev_value = statistics.stdev(data)
print("표준편차:", std_dev_value)
# 사분위수구하기
q1_value = statistics.quantiles(data, n=4)[0] # 1사분위수
q2_value = statistics.quantiles(data, n=4)[1] # 2사분위수
q3_value = statistics.quantiles(data, n=4)[2] # 3사분위수
q4_value = max(data) # 4사분위수는최대값
print("1사분위수:", q1_value)
print("2사분위수(중앙값):", q2_value)
print("3사분위수:", q3_value)
print("4사분위수:", q4_value)
# IQR (Interquartile Range) 사분위범위 구하기
iqr_value = q3_value - q1_value
print("IQR(사분위범위):", iqr_value)
```

```python
data = {'점수': [55, 60, 65, 70, 75, 80, 85, 90, 95]}
df = pd.DataFrame(data)
```

```python
# 데이터 분산성
var_value = df['점수'].var()  # Serises 데이터, var() 분산
print(var_value)
```

```python
# 표준편차
std_val = df['점수'].std() # std() 표준편차
print(std_val)
```

```python
# 사분위 점수
q1_val = df['점수'].quantile(0.25)  # Q1 하위 25%
q2_val = df['점수'].quantile(0.50)  # Q2 중앙값
q3_val = df['점수'].quantile(0.75)  # Q3 상위 25%
q4_val = df['점수'].max()           # Q4 최대값
print(q1_val, q2_val, q3_val, q4_val)
```

```python
iqr = q3_val-q1_val
print('IQR', iqr)
print(f'이상치 : {q1_val-1.5*iqr} ~ {q3_val+1.5*iqr}')
```

#### 대구에 있는 스타벅스 매장 위치 분석

```python
import pandas as pd
df = pd.read_csv('fulldata.csv', encoding="cp949", low_memory=False)
```

```python
df_1 = df[['번호','영업상태명','사업장명', '소재지전체주소','도로명전체주소']]
```

```python
# 소재지 기준 값 => NaN의 행을 삭제
df_2 = df_1.dropna(subset=['소재지전체주소'])
df_2
```

```python
# 소재지 기준 => '대구광역시' 포함하는 행만 남기는 처리!
df_3 = df_2[df_2['소재지전체주소'].str.contains('대구광역시')]
```

```python
# 사업장명이 스타벅스 인 곳
df_4 = df_3[df_3['사업장명'].str.contains('스타벅스')]
```

```python
# 영업 상태명이 폐업이 아닌 영업하는 중 인 곳
df_5 = df_4[df_4['영업상태명'].str.contains('영업')]
df_5
```

```python
'대구광역시 달서구 성당동 0261-0001 외 2필지'.split()
```

```python
# 열 추가 : 시군구 => 구별로 매장이 몇개씩 있는지!
df_5.loc[:,"시군구"] = df_5['소재지전체주소'].str.split().str[1]
df_5
```

```python
# ~별 => 그룹핑
df_6 = df_5.groupby('시군구').size().reset_index(name='사업장수')
df_6
```

```python
import matplotlib.pyplot as plt
# 한글 폰트 설정
plt.rc('font', family='Malgun Gothic') # 맑은 고딕

plt.bar(
    df_6['시군구'],
    df_6['사업장수']
)
plt.title("대구광역시 스타벅스 사업장 분석표")
plt.xlabel('시군구')
plt.ylabel('사업장수')
plt.show()
```
