```python
# 반정형 데이터 -> 정형 데이터
import re
import csv
import pandas as pd

# path
log_file_path = "./sample_logs.log"
output_csv_path = "./parsed_logs.csv"

# 정규 표현식패턴 정의
log_pattern = re.compile(
r'(?P<ip>\d+\.\d+\.\d+\.\d+)\s' # IP 주소
r'- - \[(?P<datetime>[^\]]+)\] ' # 날짜 및 시간
r'"(?P<method>GET|POST|PUT|DELETE|HEAD) ' # HTTP 메서드
r'(?P<path>[^ ]+).*?"\s' # 요청 경로
r'(?P<status>\d{3})\s' # 상태 코드
r'(?P<size>\d+)' # 응답 크기
)
# (?P<이름>패턴) 형태는"이름이있는 그룹" 으로, 추출 결과를딕셔너리처럼groupdict()로 관리할수 있게 해줍니다.
```

```python
# 결과 리스트
parsed_logs = []

# 로그 파일의 각 행의 문자열을 확인하고 정규식에 따라 결과 추출
with open(log_file_path, 'r') as log_file:
    for line in log_file:
        match = log_pattern.search(line)
        if match:
            parsed_logs.append(match.groupdict())
        
print(parsed_logs)

# CSV 파일로 저장
with open(output_csv_path, 'w', newline='') as csv_file:
    fieldnames = ['ip', 'datetime', 'method', 'path', 'status', 'size']
    writer = csv.DictWriter(csv_file, fieldnames=fieldnames)
    writer.writeheader()
    for log in parsed_logs:
        writer.writerow(log)

print("보안 로그가 CSV 파일로 저장되었습니다:", output_csv_path)
```

```python
# IQR(Interquartile Range) 기반 이상치 처리
import numpy as np
# 이상치 처리
data = {'Scores': [70, 75, 80, 85, 90, 150]}
df = pd.DataFrame(data)

Q1 = df['Scores'].quantile(0.25)  # Q1
Q3 = df['Scores'].quantile(0.75)  # Q3
IQR = Q3 - Q1  # IQR 계산
#print(Q1, Q3, IQR)

low_bound = Q1 - 1.5 * IQR
high_bound = Q3 + 1.5 * IQR
#print(low_bound, high_bound)

#df_result = df[(df['Scores'] >= low_bound) & (df['Scores'] <= high_bound)]
# 이상치 데이터의 변경
df['Scores'] = df['Scores'].astype(float)  # 데이터 타입을 float으로 변환
df.loc[df['Scores'] < low_bound, 'Scores'] = low_bound
df.loc[df['Scores'] > high_bound, 'Scores'] = high_bound
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler, LabelEncoder, OneHotEncoder

# 데이터 생성
data = pd.DataFrame({'Original': [50, 80, 90, 100, 60]})

# 표준화
scaler1 = StandardScaler()
data['scaled_1'] = scaler1.fit_transform(data[['Original']])

# 정규화
scaler2 = MinMaxScaler()
data['scaled_2'] = scaler2.fit_transform(data[['Original']])

data
```

```python
data = pd.DataFrame({'Fruits': ['사과', '배', '포도', '사과', '포도']})

encoder1 = LabelEncoder()
data['Label_Encoded'] = encoder1.fit_transform(data['Fruits'])
data
```

```python
# sparse_output=False는 결과를 희소 행렬이 아닌 일반 배열로 반환하도록 설정하는 옵션입니다.
onehot_encoder = OneHotEncoder(sparse_output=False) 

onehot_encoded = onehot_encoder.fit_transform(data[['Label_Encoded']])
onehot_encoded
```

```python
onehot_columns = onehot_encoder.get_feature_names_out()
onehot_df = pd.DataFrame(onehot_encoded, columns=onehot_columns)
print(onehot_columns)
print(onehot_df)
```

```python
# axis=1은 열 방향으로 결합을 의미합니다. 즉, 기존 데이터프레임(data)과 
# 원-핫 인코딩된 데이터프레임(onehot_df)을 열 단위로 합쳐서 
# 새로운 데이터프레임(result_df)을 생성합니다.
result_df = pd.concat([data, onehot_df], axis=1)
result_df
```
