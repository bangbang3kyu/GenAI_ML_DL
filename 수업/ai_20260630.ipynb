```python
import requests as req

# 요청 url
url ='https://jsonplaceholder.typicode.com/posts'

# 요청 : read => get()
res = req.get(url)
print(res,type(res))
```

```python
# http 상태 코드 => status_code
if res.status_code == 200:
    print('OK!!')
    print(res.text[:200])
else :
    print('Error!!!', res.status_code)

# JSON(text) => Object
data_obj = res.json()
```

```python
from bs4 import BeautifulSoup as bs 
html_doc = """
<html>
    <head>
        <title>Example Page</title>
    </head>
    <body>
        <h1>Welcome to Web Scraping</h1>
        <p class="description a b big">Welcome to Web Scraping.</p>
        <a href="https://example.com1">Visit Example1</a>
        <a href="https://example.com2">Visit Example2</a>
        <a href="https://example.com3">Visit Example3</a>
        <a href="https://example.com4" id="first">Visit Example4</a>
    </body>
</html>
"""
# BeautifulSoup 객체 생성
soup = bs(html_doc, 'html.parser')

# 태그 탐색
print(soup.title.string)
print(soup.h1.string)
print(soup.select('#first'))
```

```python
# find(tag), find_all(tag)
a_tag = soup.find('a')
print(a_tag)
a_tags = soup.find_all('a')
print(a_tags)
```

```python
links = []
for a_tag in a_tags:
    links.append(a_tag['href'])

print(links)
```

```python
# select(selector) => list
# select_one(selector) => tag
# class="description title big small"
# .description {} .title {} .big {} .small {}
# id="first" => #first
# tag 선택
# 모든 태그 => *
# 클래스가 "description"인 <p> 태그 찾기
description = soup.select_one('.description')
print(description)
```

```python
# select(selector) => list
# select_one(selector) => tag
# class="description title big small"
# .description {} .title {} .big {} .small {}
# id="first" => #first
# tag 선택
# 모든 태그 => *
tag1 = soup.select_one('.description')
print(tag1)
tag2 = soup.select('.description')
print(tag2)
```

#### 관계형 데이터 베이스 시스템 : RDBMS => SQLite

```python
import sqlite3
```

```python
# 데이터 베이스 연결
conn = sqlite3.connect('first.db')
print('데이터베이스 연결 성공!')
conn.close()
```

```python
# 관계형 데이터 베이스의 특징 => 테이블 형식으로 데이터 관리
# 데이터를 저장할 공간을 생성 => 테이블 생성
sql_create = '''
CREATE TABLE IF NOT EXISTS users (
id INTEGER PRIMARY KEY AUTOINCREMENT,
name TEXT NOT NULL,
age INTEGER,
city TEXT
)
'''
```

```python
# 1. 데이터베이스 연결 ⇒ connection
conn = sqlite3.connect('first.db')
print('connn... ok!!!')
```

```python
# 2. sql 실행 객체 ⇒ cursor 객체 구하기
cursor = conn.cursor()
```

```python
# 3. execute (SQL 실행시키는 매개체)
cursor.execute(sql_create)
```

```python
# 4. close
conn.close()
```

```python
# 데이터 입력 => 행 단위 입력
# id, name, age, city
# sql
sql_insert = 'insert into users (name, age, city) values (?,?,?)'

# 데이터베이스 연결
conn = sqlite3.connect('first.db')

data = [
    ('hot', 21, 'seoul'),
    ('king', 30, 'jeju'),
    ('son', 25, 'LA')
]

# Cursor
cursor = conn.cursor()

# sql 실행
#cursor.execute(sql_insert, ('cool',20,'daegu'))
cursor.executemany(sql_insert,data[:])

# 작업 완료 : 입력, 수정, 삭제 => 완료 또는 초기화
conn.commit()

# close
conn.close()
```

```python
# 행단위 검색: 질의 
sql_select = 'select * from users'
sql_select2 = 'select * from users where age>25'

conn = sqlite3.connect('first.db')

cursor = conn.cursor()

cursor.execute(sql_select2)

# select 결과 => 새로운 테이블
# fetchall() => 전체 행 반환
# fetchone() => 하나의 행 반환  
rows = cursor.fetchall()
#print(rows)

for row in rows:
    print(row)

conn.close()
```

```python
# 데이터 수정
# 행을 찾고 => 특정 컬럼의 데이터를 수정
sql_update = 'update users set age=?, name=? where id=?'

conn = sqlite3.connect('first.db')

cursor = conn.cursor()

# data = [
#     (15, 'Park',1),

# ]

cursor.execute(sql_update, (15, 'Park', 1))

conn.commit()

conn.close()
```

```python
# 삭제 => 행단위 삭제
sql_delete ='delete from users where id=?'

conn = sqlite3.connect('first.db')

cursor = conn.cursor()

cursor.execute(sql_delete, (1,))

conn.commit()

conn.close()
```
